# Find the Most Expensive Apps, Filter out the Junk, and Calculate a (ballpark) Sales Revenue Estimate

Let's examine the Price column more closely.

Convert the price column to numeric data. Then investigate the top 20 most expensive apps in the dataset.

Remove all apps that cost more than $250 from the `df_apps_clean` DataFrame.

Add a column called 'Revenue_Estimate' to the DataFrame. This column should hold the price of the app times the number of installs. What are the top 10 highest grossing paid apps according to this estimate? Out of the top 10 highest grossing paid apps, how many are games?


```python
df_apps_clean2 = df_apps_clean.copy()
df_apps_clean2['Price'] = pd.to_numeric(df_apps_clean2['Price'].str.strip('$'))
df_apps_clean2[['App','Price']].sort_values(by='Price', ascending=False).head(20)
```

### The most expensive apps sub $250

```python
# Remove all apps that cost more than $250 from the df_apps_clean DataFrame.
df_apps_clean3 = df_apps_clean2.copy()
df_apps_clean3 = df_apps_clean2[df_apps_clean2['Price'] <= 250]
df_apps_clean3.sort_values(by='Price', ascending=False)
```

### Highest Grossing Paid Apps (ballpark estimate)

```python
# What are the top 10 highest-grossing paid apps according to this estimate? Out of the top 10, how many are games?
df_apps_clean3 = df_apps_clean3.copy()
df_apps_clean3['Revenue_Estimate'] = df_apps_clean3['Price'] * df_apps_clean3['Installs']
df_apps_clean3
df_apps_clean3[['App', 'Category', 'Installs', 'Price', 'Revenue_Estimate']].sort_values(by='Revenue_Estimate', ascending=False).head(10)
```

```python
df_apps_clean3.Category.nunique()
top10_category = df_apps_clean3.Category.value_counts()[:10]
top10_category
```

# Plotly Bar Charts & Scatter Plots: Analysing App Categories

```python
# Let's visualize not just the total number of apps in the category but how often apps are downloaded in that category
category_installs = df_apps_clean3.groupby('Category').agg({'Installs': pd.Series.sum})
category_installs.sort_values('Installs', ascending=True, inplace=True)
category_installs

```

```python
h_bar = px.bar(x=category_installs.Installs,
               y=category_installs.index,
               orientation='h',
               title='Category Popularity')

h_bar.update_layout(xaxis_title='Number of Downloads',
                    yaxis_title='Category')

h_bar.show()
```

```python
df_category_vs_apps = df_apps_clean3.groupby('Category').agg({
    'App': 'count',
    'Installs': 'sum'
})

df_category_vs_apps.sort_values('Installs', ascending=False, inplace=True)
df_category_vs_apps

# Another way:
# cat_number = df_apps_clean.groupby('Category').agg({'App': pd.Series.count})
# cat_merged_df = pd.merge(cat_number, category_installs, on='Category', how="inner")
# print(f'The dimensions of the DataFrame are: {cat_merged_df.shape}')
# cat_merged_df.sort_values('Installs', ascending=False)
```

```python
scatter = px.scatter(df_category_vs_apps, # data
                    x='App', # column name
                    y='Installs',
                    title='Category Concentration',
                    size='App',
                    hover_name=df_category_vs_apps.index,
                    color='Installs')

scatter.update_layout(xaxis_title="Number of Apps (Lower=More Concentrated)",
                      yaxis_title="Installs",
                      yaxis=dict(type='log'))

scatter.show()
```

# Extracting Nested Data from a Column

How many different types of genres are there? Can an app belong to more than one genre?


```python
df_apps_multiple_genres = df_apps_clean[df_apps_clean["Genres"].str.contains("Education")]
df_apps_multiple_genres
```

```python
df_apps_multiple_genres.Genres.value_counts()
```

```python
separated_genres_df = df_apps_multiple_genres.Genres.str.split(";")
separated_genres_df
```

```python
# Using expand to have values in different columns instead of same
separated_genres_df = df_apps_multiple_genres.Genres.str.split(";", expand=True)
separated_genres_df
```

```python
# Use stack to put column 1 belown column 0 to calculate more efficiently
stacked_genres = separated_genres_df.stack()
stacked_genres
```

```python
# Get the count by Individual Genre
stacked_genres.value_counts()
```

```python
# Split the strings on the semi-colon and then .stack them.
stack = df_apps_clean3.Genres.str.split(';', expand=True).stack()
stack
print(f'We now have a single column with shape: {stack.shape}')
num_genres = stack.value_counts()
num_genres
print(f'Number of genres: {len(num_genres)}')
```

```python
num_genres
```

# Plotly Box Plots: Lost Downloads for Paid Apps

Create a box plot that shows the number of Installs for free versus paid apps. How does the median number of installations compare? Is the difference large or small?



```python
box = px.box(
    df_apps_clean3,
    x='Type',
    y='Installs',
    title='How Many Downloads Are Paid Apps Giving Up?',
    color='Type',
    notched=True,
    boxmode='group',
    points="all"
)

box.update_layout(xaxis_title='Type',
                    yaxis_title='Installs',
                    yaxis=dict(type='log'))

box.show()
```
